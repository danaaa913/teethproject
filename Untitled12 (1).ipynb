```python
import os
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense

#  إعداد ImageDataGenerator لتحميل البيانات مع التطبيع
train_data_gen = ImageDataGenerator(rescale=1./255)
validation_data_gen = ImageDataGenerator(rescale=1./255)
test_data_gen = ImageDataGenerator(rescale=1./255)

#  تحميل بيانات التدريب
train_data = train_data_gen.flow_from_directory(
    '/content/drive/MyDrive/Colab Notebooks/Teeth_Dataset/Training',
    target_size=(256, 256),
    batch_size=32,
    class_mode='categorical'  # تصنيف متعدد الفئات
)

#  تحميل بيانات التحقق
validation_data = validation_data_gen.flow_from_directory(
    '/content/drive/MyDrive/Colab Notebooks/Teeth_Dataset/Validation',
    target_size=(256, 256),
    batch_size=32,
    class_mode='categorical'  # تصنيف متعدد الفئات
)

#  تحميل بيانات الاختبار
test_data = test_data_gen.flow_from_directory(
    '/content/drive/MyDrive/Colab Notebooks/Teeth_Dataset/Testing',
    target_size=(256, 256),
    batch_size=32,
    class_mode='categorical'  # تصنيف متعدد الفئات
)

#  بناء نموذج CNN
model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(256, 256, 3)),
    MaxPooling2D(2, 2),
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D(2, 2),
    Flatten(),
    Dense(128, activation='relu'),
    Dense(7, activation='softmax')  # 7 فئات
])

#  تجميع النموذج
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

#  تدريب النموذج
history = model.fit(train_data, epochs=10, validation_data=validation_data)

# تقييم النموذجر
test_loss, test_acc = model.evaluate(test_data)
print(f"Test accuracy: {test_acc}")

#  رسم دقة التدريب والخسارة
import matplotlib.pyplot as plt

# رسم دقة النموذج
plt.plot(history.history['accuracy'], label='training accuracy')
plt.plot(history.history['val_accuracy'], label='validation accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.show()

# رسم خسارة النموذج
plt.plot(history.history['loss'], label='training loss')
plt.plot(history.history['val_loss'], label='validation loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```
